```python
# Import libraries 
import matplotlib.pyplot as plt     #used for plotting data 
import numpy as np     #used for mathematical operations
import pandas as pd    #used to loading CSV data
```

# Bar Chart


```python
donuts_per_person = {'Kwanda': 6, 'Jonathan': 8, 'Jaco': 7, 'Jason': 5, 'Laure': 6, 'Andrew': 8 }

donuts = donuts_per_person.values()
people = donuts_per_person.keys()

y_pos = np.arange(len(donuts_per_person))    # a range of number which is the length of donuts_per_person dict, i.e 6 now

# To set up the graph
plt.bar(y_pos, donuts, color= 'blue', alpha= 0.55)    # y_pos is number of categories, num of donuts for each categ., alpha- transparency or color of each bar 90%
plt.xticks(y_pos, people)    #each of the markers going along x-axis, setting it to no. of items in y_pos, values will be people
plt.ylabel('Donuts')    # label of the y-axis
plt.title('How many donuts each')    # title of plot

#To show it
plt.show()
```

# Pie Chart

```python
labels = donuts_per_person.keys()    # keys - names of each of the people in the dict
sizes = donuts_per_person.values()   # values - size of each, no. of donuts eaten by each person
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'grey', 'blue']    # we know they are 6, so 6 colors


# Plot
"""
labels = our labels, which are the keys in the dictionary
autopct = automatically shows numbers/size of each segment as percentages, with a float value with 1 decimal place
startangle = we didnt wwant it to start with a vertical line at the top, so we just gave it 140 degrees
"""
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140 )
plt.axis('equal')    # equal ensures that the pie chart is a circle and not an eclipse or oval or something
plt.show()
```

# Line Graph
#### WIth Time Series Data

```python
total_donuts = { '31/01/2019': 40, '28/02/2019': 38, '31/03/2019': 30, '30/04/2019': 28, '31/05/2019': 25, '30/06/2019': 35}

dates = list(total_donuts.keys())    # convert the dict to a list, keys will be our dates
x_ax = [date[3:5] for date in dates]    #we only want characters in position 3 and 4- the MONTH, of each date
y_ax = list(total_donuts.values())

plt.plot(x_ax, y_ax)
plt.xlabel('Month')
plt.ylabel('Number of Total Donuts Eaten')
plt.title('Line Plot showing the Number of Donuts Eaten over the First 6 Months of 2019')

```

### Analysis
##### Number of donuts eaten decreases from Jan to May, when winter starts in SA, after May, when winter starts, there's a significant uptake
##### in donuts. AND that's the same time when Jonathan and  Andrew started working at the academy

# Scatter Plot

```python
#key = donuts eaten
# values = age
donuts_per_age = {10:25, 16:22, 15:20, 12:21, 5:30, 6:31, 10:29, 2:35, 4:32, 8:29, 7:29, 2:30, 8:31, 18:22, 18:24, 19:25}

x_ax = donuts_per_age.values()
y_ax = donuts_per_age.keys()

#Plot
plt.scatter(x_ax, y_ax, alpha = 0.9)
plt.title('Scatter Plot showing number of Donuts eaten per Age')
plt.xlabel('Age')
plt.ylabel('Number of Donuts Eaten')
plt.show()
```

# Analysis
#####  General trend is that older people tend to eat less donuts, implying a negative correlation between age and donut consumption. However, this is not a certainty, as our analysis is only preliminary, but it gives us a good place to start.
